P(A & B): the probaility a sentences/txt relevant to a subtopics and also get covered by the main topic

P(A & B) = P(A) *  P(B) 

```python
txt_subtopics_extracted_eval_rank = txt_subtopics_extracted_eval_df[["main_topic", "Title", "subtopic_cv_ratio"]].drop_duplicates()

txt_subtopics_extracted_eval_rank
```

### Total Video File Txt level main topic relevance score: 
    
take the dot prudtc between the the vector of the extracted subtopic probaility with the vector of subtpics count ratio agins the total numbers of trained subtopics

- txt_main_topic_classified_proba: used to measure the probaility of a main topics covers the transcript text

- txt_subtopics_proba: used to measure a single subtopic relevance quality extracted from a transcript text; we main use this KPI to rank the video files subtopic relevance quality under a given transcript text


- total_txt_level_subtopics_relevance_score: is the score encaptuate and measure the total sub-topic extraction quality of a text transcript.

      total_txt_level_subtopics_relevance_score = dot(subtopic_proba, extracted_subtopic_count_ratio) 
      
- total_txt_level_relevance_score: is the score measure the quality by a total relavance score from both the main topics and their extracted subtopics; we main use this KPI to rank the video files overall relevance quality under a given main topics 

      weighted_total_txt_level_relevance_score = total_txt_level_subtopics_relevance_score * main_topic_classified_proba
      
      total_txt_level_relevance_score = txt_main_topic_classified_proba * main_topic_classified_proba

- total_txt_level_subtopic_level_relevance_score: measure the relavnce quality of a indvidual subtopic extracted from a transcript text and the proba of getiing covered by the main topic; we main use this KPI to rank the video files subtopic total relevance quality under a given main topics 

      txt_level_subtopics_level_relevance_score = subtopic_proba * main_topic_classified_proba

```python
def total_subtopic_probab_score(DF, fl_nm,  proba_col, cnt_ratio, stences = None, txt_lvl = True):
    
    if txt_lvl == True:

        temp_DF = DF[DF['Title'] == fl_nm]
        
#         print(temp_DF)

        return np.round(np.dot(np.array(list(temp_DF[proba_col])), \
               np.array(list(temp_DF[cnt_ratio]))), 4)
    else: # sentences 
        
        temp_DF = DF[(DF['Title'] == fl_nm) & (DF['sentences'] == stences)]

        return np.round(np.dot(np.array(list(temp_DF[proba_col])), \
               np.array(list(temp_DF[cnt_ratio]))), 4)
```

```python
txt_lvl_total_relevance_calcu = text_level_relevance_score_df[['Title', 'main_topic', 'txt_subtopics_indx', \
                                'txt_subtopics_proba', 'subtopic_cv_ratio_x', 'txt_main_topic_classified_proba']].drop_duplicates()

txt_lvl_total_relevance_calcu.head()
```

```python
# calculate the total_txt_level_subtopics_relevance_score

file_names = list(txt_lvl_total_relevance_calcu['Title'].unique())
total_subtopic_relevance_scores = []

for file in file_names:
#     print(file)
    total_rel_score = total_subtopic_probab_score(\
                        DF = txt_lvl_total_relevance_calcu, fl_nm = file, \
                    proba_col = 'txt_subtopics_proba', cnt_ratio='subtopic_cv_ratio_x')
    total_subtopic_relevance_scores.append(total_rel_score)
#     print(total_rel_score)
    
txt_lvl_relevance_scores_df = pd.DataFrame({"Title" : file_names
            ,"total_txt_level_subtopics_relevance_score": total_subtopic_relevance_scores, 
             })
txt_lvl_relevance_scores_df.head()
```

```python
# calculate total_txt_level_relevance_score
txt_lvl_total_relevance_calcu = pd.merge(txt_lvl_total_relevance_calcu, txt_lvl_relevance_scores_df, "left", \
         left_on = ['Title'], right_on = ['Title'])
txt_lvl_total_relevance_calcu['total_txt_level_relevance_score'] = txt_lvl_total_relevance_calcu['total_txt_level_subtopics_relevance_score'] * \
                                          txt_lvl_total_relevance_calcu['txt_main_topic_classified_proba'] 
txt_lvl_total_relevance_calcu.head()
```

##### given 5 main topics what are the top 10 video transcript txt in terms of total relavnce quality with total_txt_level_relevance_score

```python
main_topics_lst = list(txt_lvl_total_relevance_calcu['main_topic'].unique())

for main_topic in main_topics_lst:
    print()
    print('--------------------------')
    print("Under a main_topic {},  we have our top 10 videos ranked by total_relevance_score: \n".format(main_topic))
    [print(i) for i in list(txt_lvl_total_relevance_calcu[txt_lvl_total_relevance_calcu['main_topic'] ==\
                                  main_topic][['Title', 'main_topic', \
                                'total_txt_level_relevance_score']].drop_duplicates().sort_values(\
                                    by = 'total_txt_level_relevance_score', ascending = False).head(10)['Title'])]
```
